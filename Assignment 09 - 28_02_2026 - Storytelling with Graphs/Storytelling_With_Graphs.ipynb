```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("titanic.csv")

# --- Bar Chart: Survival Count ---
survival_counts = df["Survived"].value_counts()

plt.figure()
survival_counts.plot(kind="bar")
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")
plt.show()


# --- Pie Chart: Gender Distribution ---
gender_counts = df["Sex"].value_counts()

plt.figure()
gender_counts.plot(kind="pie", autopct='%1.1f%%')
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()


# --- Histogram: Age Distribution ---
plt.figure()
df["Age"].dropna().plot(kind="hist", bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

The analysis of the Titanic dataset reveals several interesting patterns. The bar chart shows that a majority of passengers did not survive the disaster, highlighting the severity of the event. The pie chart indicates that there were significantly more male passengers than female passengers on board.

The histogram of age distribution shows that most passengers were between 20 and 40 years old, with fewer elderly passengers and children.

Combining these insights, it can be inferred that survival was not evenly distributed and may have been influenced by factors such as gender and age. Historical trends suggest that women and children were given priority during evacuation, which aligns with the observed patterns in the dataset.
