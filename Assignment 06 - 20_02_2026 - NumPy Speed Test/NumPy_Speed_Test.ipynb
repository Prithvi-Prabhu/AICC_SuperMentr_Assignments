**Objective**

To compare the performance of Python lists and NumPy arrays when handling 1 million numbers by measuring execution time.

```python
import time
import numpy as np

# Create data
size = 1_000_000

# Python list
py_list = list(range(size))

# NumPy array
np_array = np.arange(size)

# ---- Operation: Add 5 to each element ----

# Python list timing
start = time.time()
py_result = [x + 5 for x in py_list]
end = time.time()
print(f"Python list time: {end - start:.5f} seconds")

# NumPy array timing
start = time.time()
np_result = np_array + 5
end = time.time()
print(f"NumPy array time: {end - start:.5f} seconds")
```

**Observations**

Performance Difference
NumPy arrays are significantly faster than Python lists because operations are executed in optimized C code rather than Python loops.

Memory Efficiency
NumPy arrays consume less memory since they store elements of the same data type, while Python lists store references to objects.

Vectorized Operations
NumPy supports vectorized operations (e.g., array + 5), eliminating the need for explicit loops, making computations faster and more concise.

**Conclusion**

NumPy arrays are more efficient than Python lists for large-scale numerical computations due to faster execution, lower memory usage, and support for vectorized operations.
