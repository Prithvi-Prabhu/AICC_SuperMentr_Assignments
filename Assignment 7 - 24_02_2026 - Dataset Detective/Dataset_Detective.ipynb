```python
import pandas as pd

def analyze_dataset():
    # Load dataset
    df = pd.read_csv("titanic.csv")

    # Display top rows
    print("\n--- Top 5 Rows ---")
    print(df.head())

    # Find column with highest value
    numeric_cols = df.select_dtypes(include=['number'])
    max_values = numeric_cols.max()
    highest_column = max_values.idxmax()

    print("\n--- Column with Highest Value ---")
    print(f"{highest_column} (Max Value: {max_values[highest_column]})")

    # Count missing values
    print("\n--- Missing Values Count ---")
    print(df.isnull().sum())

    # Insights
    print("\n--- Insights ---")

    # 1. Dataset shape
    print(f"1. Dataset has {df.shape[0]} rows and {df.shape[1]} columns.")

    # 2. Survival rate
    survival_rate = df["Survived"].mean() * 100
    print(f"2. Survival Rate: {survival_rate:.2f}%")

    # 3. Average age
    avg_age = df["Age"].mean()
    print(f"3. Average Age: {avg_age:.2f}")

    # 4. Gender survival comparison
    gender_survival = df.groupby("Sex")["Survived"].mean()
    print("4. Survival rate by gender:")
    print(gender_survival)

    # 5. Class survival comparison
    class_survival = df.groupby("Pclass")["Survived"].mean()
    print("5. Survival rate by passenger class:")
    print(class_survival)


analyze_dataset()
```

**Insights:**

The dataset contains information about passengers such as age, gender, class, and survival status.

The overall survival rate is around 38%, indicating most passengers did not survive.

Females had a significantly higher survival rate compared to males.

Passengers in 1st class had a higher survival rate than those in lower classes.

The dataset contains missing values, especially in the Age and Cabin columns, which may require data cleaning.
