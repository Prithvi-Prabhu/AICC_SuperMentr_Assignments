```python
# Text Cleaner Program

import re

# -------------------------------
# 1. Define Stopwords
# -------------------------------
# Basic stopwords list (can be expanded)
stopwords = {
    "is", "the", "and", "a", "an", "in", "on", "at", "to", "for",
    "of", "with", "this", "that", "it", "be", "as", "are"
}

# -------------------------------
# 2. Text Cleaning Function
# -------------------------------
def clean_text(text):

    # Convert to lowercase
    text = text.lower()

    # Remove punctuation (anything not letter/number/space)
    text = re.sub(r"[^\w\s]", "", text)

    # Tokenize (split into words)
    words = text.split()

    # Remove stopwords
    cleaned_words = [word for word in words if word not in stopwords]

    # Join back into string
    return " ".join(cleaned_words)

# -------------------------------
# 3. Test the Cleaner
# -------------------------------
test_sentence = "This is an AMAZING product! It works perfectly, and I love it."

cleaned = clean_text(test_sentence)

print("Original Text:")
print(test_sentence)

print("\nCleaned Text:")
print(cleaned)
```
