```python
# Simple Movie Review Sentiment Analyzer

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# -------------------------------
# 1. Training Data
# -------------------------------
reviews = [
    "I loved this movie",
    "This film was amazing",
    "Fantastic acting and story",
    "I hated this movie",
    "Worst movie ever",
    "Terrible experience"
]

labels = [1, 1, 1, 0, 0, 0]  # 1 = Positive, 0 = Negative

# -------------------------------
# 2. Convert Text to Features
# -------------------------------
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(reviews)

# -------------------------------
# 3. Train Model
# -------------------------------
model = MultinomialNB()
model.fit(X, labels)

# -------------------------------
# 4. Test on 5 New Reviews
# -------------------------------
test_reviews = [
    "Amazing movie with great acting",
    "I really hated the film",
    "It was a fantastic experience",
    "Worst acting ever",
    "I enjoyed the movie"
]

X_test = vectorizer.transform(test_reviews)
predictions = model.predict(X_test)

# -------------------------------
# 5. Output Results
# -------------------------------
print("\n--- Sentiment Predictions ---")

for review, pred in zip(test_reviews, predictions):
    sentiment = "Positive" if pred == 1 else "Negative"
    print(f"{review} --> {sentiment}")
```
