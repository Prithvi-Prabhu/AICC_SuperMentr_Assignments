```python
import pandas as pd

def clean_dataset():
    # Load dataset
    df = pd.read_csv("titanic.csv")

    print("\n--- Original Dataset Info ---")
    print(df.info())

    # 1. Handle missing values
    # Fill Age with mean
    df["Age"].fillna(df["Age"].mean(), inplace=True)

    # Fill Embarked with mode
    df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)

    # Drop Cabin column (too many missing values)
    df.drop(columns=["Cabin"], inplace=True)

    # 2. Remove duplicates
    df.drop_duplicates(inplace=True)

    # 3. Standardize text
    df["Sex"] = df["Sex"].str.lower().str.strip()
    df["Embarked"] = df["Embarked"].str.upper().str.strip()

    print("\n--- Cleaned Dataset Info ---")
    print(df.info())

    print("\n--- Sample Cleaned Data ---")
    print(df.head())

clean_dataset()
```

**Steps Performed**

*Handled Missing Values*

Filled missing Age values with the mean

Filled missing Embarked values with the most frequent value

Dropped Cabin column due to excessive missing data

*Removed Duplicates*

Ensured no repeated rows exist in the dataset

*Standardized Text*

Converted text to consistent formats (lowercase/uppercase)
Removed unnecessary spaces

**Why Data Cleaning Matters**

*Improves Accuracy*

Clean data ensures machine learning models and analysis produce correct results.

*Handles Missing Data Issues*

Missing values can lead to errors or biased outcomes if not treated properly.

*Ensures Consistency*

Standardized text avoids confusion (e.g., "Male", "male", "MALE").

*Removes Noise*

Duplicate or incorrect data can distort analysis and insights.

*Enhances Performance*

Clean datasets make computations faster and more efficient.

**Conclusion**

Data cleaning is a crucial step in data analysis. Without it, insights can be misleading and models unreliable. Proper preprocessing ensures high-quality, trustworthy results.
