This assignment demonstrates how to design a dataset, identify features and labels, and model the relationship using machine learning. It also includes visualization and model evaluation to better understand the data.

**DATASET CREATION**

```python
import pandas as pd

# Create dataset
data = {
    "Study Hours": [1, 2, 3, 4, 5, 6, 7, 8, 2.5, 3.5, 6.5, 7.5],
    "Marks": [30, 35, 50, 55, 65, 70, 80, 90, 40, 52, 75, 85]
}

df = pd.DataFrame(data)

print("\nDataset:")
print(df)
```

**Feature and Label Identification**

*   Feature (Independent Variable): Study Hours
*   Label (Dependent Variable): Marks

The feature represents the input used for prediction, while the label is the output the model learns to predict.

**EXPLORATORY DATA ANALYSIS (EDA)**

```python
print("\nBasic Statistics:")
print(df.describe())

print("\nCorrelation:")
print(df.corr())
```

**Insight:**

A strong positive correlation between study hours and marks indicates that as study time increases, marks tend to increase.

**MODEL BUILDING - LINEAR REGRESSION**

```python
from sklearn.linear_model import LinearRegression

X = df[["Study Hours"]]
y = df["Marks"]

model = LinearRegression()
model.fit(X, y)
```

**PREDICTION**

```python
predicted_marks = model.predict([[9]])
print(f"\nPredicted marks for 9 study hours: {predicted_marks[0]:.2f}")
```

**MODEL EVALUALTION**

```python
from sklearn.metrics import r2_score

y_pred = model.predict(X)
r2 = r2_score(y, y_pred)

print(f"R² Score: {r2:.2f}")
```

**Interpretation:**

The R² score indicates how well the model explains the variance in marks. A value close to 1 suggests a strong relationship.

**VISUALIZATION**

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks (Regression Line)")
plt.show()
```

**Relationship Analysis**

The dataset demonstrates a strong positive linear relationship between study hours and marks. The regression model effectively captures this trend, indicating that increased study time leads to improved academic performance.
