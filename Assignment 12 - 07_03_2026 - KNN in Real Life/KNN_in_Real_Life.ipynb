This assignment explains how the K-Nearest Neighbors (KNN) algorithm is used in recommendation systems (like Netflix) and demonstrates a simple similarity-based example.

**Understanding KNN in Real Life**

K-Nearest Neighbors (KNN) is a simple machine learning algorithm that makes predictions based on the similarity between data points. It finds the “K” most similar items (neighbors) and uses them to make decisions.

In platforms like Netflix, KNN is used to recommend movies or shows by identifying users with similar preferences.

**How Recommendation Works (Netflix-like)**

Each user has a history of watched/rated movies
The system compares users based on similarity (e.g., ratings)
It finds the K most similar users
Recommends movies that those similar users liked

This is called a user-based collaborative filtering approach

**Simple Similarity Example**

Dataset:

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Sample user-movie ratings
data = {
    "Movie A": [5, 4, 1],
    "Movie B": [4, 5, 1],
    "Movie C": [1, 2, 5]
}

users = ["User1", "User2", "User3"]

df = pd.DataFrame(data, index=users)
print("User Ratings:\n", df)
```

```python
similarity = cosine_similarity(df)

similarity_df = pd.DataFrame(similarity, index=users, columns=users)
print("\nUser Similarity Matrix:\n", similarity_df)

# Find most similar user to User1
most_similar = similarity_df["User1"].sort_values(ascending=False)[1]
print(f"\nMost similar to User1: {most_similar}")
```

**Recommendation Logic**

If User1 hasn’t watched a movie that their nearest neighbor liked, recommend it.

**Output Explanation**

Users with similar ratings have higher similarity scores
The system identifies nearest neighbors
Recommendations are based on neighbor preferences

**Key Concepts Used**

KNN (distance/similarity-based learning)
Cosine similarity
Collaborative filtering
