```python
import re

# Example sentence
text = "omg this movie is sooo goood!!! 😍🔥"

# Step 1: Lowercase
text = text.lower()

# Step 2: Remove URLs
text = re.sub(r"http\S+", "", text)

# Step 3: Remove emojis (basic)
text = re.sub(r"[^\w\s]", "", text)

# Step 4: Normalize repeated characters
text = re.sub(r"(.)\1{2,}", r"\1\1", text)

# Step 5: Replace slang
slang_dict = {
    "omg": "oh my god",
    "u": "you",
    "thx": "thanks",
    "luv": "love",
    "rn": "right now"
}

words = text.split()
words = [slang_dict.get(word, word) for word in words]
text = " ".join(words)

print("Cleaned Text:", text)
```
