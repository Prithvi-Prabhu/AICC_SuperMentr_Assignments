```python
# Word Importance Explorer using TF-IDF

from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# -------------------------------
# 1. Documents
# -------------------------------
documents = [
    """Artificial Intelligence is transforming industries by enabling machines to learn from data.
    It is widely used in healthcare, finance, and automation to improve efficiency and decision-making.""",

    """Cricket is one of the most popular sports in the world, especially in countries like India.
    Players require skill, teamwork, and strategy to perform well in matches and tournaments.""",

    """Healthy eating and regular exercise are essential for maintaining good health.
    A balanced diet provides necessary nutrients and helps prevent diseases and improves overall wellbeing.""",

    """E-commerce platforms like online shopping websites allow users to purchase products easily.
    Recommendation systems help suggest products based on user preferences and browsing history.""",

    """Education is evolving with the use of technology such as online learning platforms.
    Students can access courses remotely and improve their skills through digital resources."""
]

# -------------------------------
# 2. Apply TF-IDF
# -------------------------------
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(documents)

# Convert to DataFrame
df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

print("\n--- TF-IDF Matrix ---")
print(df)

# -------------------------------
# 3. Extract Top Keywords
# -------------------------------
print("\n--- Top Keywords per Document ---")

for i in range(len(documents)):
    top_words = df.iloc[i].sort_values(ascending=False).head(5)

    print(f"\nDocument {i+1}:")
    print(top_words)
```
